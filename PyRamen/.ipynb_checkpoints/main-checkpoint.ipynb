```python
# Import the os and csv library
import os
import csv


# Get csv file paths for menu_data and sales_data
csvpath_menu = os.getcwd()+'\\Resources\\menu_data.csv'
csvpath_sales = os.getcwd()+'\\Resources\\sales_data.csv'

# Initialize lists to hold data from menu_data.csv and sales.csv
# menu list
menu_items = []
price = []
cost = []

# sales list
quantity = []
sales_items = []


# Import data from menu_data.csv into menu_items, price and cost lists

with open(csvpath_menu, 'r') as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    header = next(csvreader)
    for row in csvreader:
        item = row[0]
        revenue = float(row[3])
        cogs = float(row[4])
        
        menu_items.append(item)
        price.append(revenue)
        cost.append(cogs)
          

# Import data from sales_data.csv into quantity and sales_items lists

with open(csvpath_sales, 'r') as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    header = next(csvreader)
    for row in csvreader:
        item = row[4]
        qty = int(row[3])
        
        sales_items.append(item)
        quantity.append(qty)

        
## Perform calculations

# Calculate profit margin
        
margin = []
for i in range(0, len(menu_items)):
    profit = price[i] - cost[i]
    margin.append(profit)
    

# Calculate total quantit of menu items
    
total_item_quantity = []

for menu in menu_items:
    
    total_quantity = 0
    
    for i in range(0, len(sales_items)):
        if (sales_items[i] == menu):
            total_quantity += quantity[i]
    total_item_quantity.append(total_quantity)
    

# Create empty report dictionary to hold menu items details
report = {}


''' 
    Next we iterate through menu items and for those items with quantity greater than 0, we calculate the 
    items details and append to the report dictionary

'''
for i in range(0,len(menu_items)):
    if(total_item_quantity[i] > 0):
        sales_item = menu_items[i]
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}

        report[sales_item]["01-count"] = total_item_quantity[i]
        report[sales_item]["02-revenue"] = total_item_quantity[i] * price[i]
        report[sales_item]["03-cogs"] = total_item_quantity[i] * cost[i]
        report[sales_item]["04-profit"] = total_item_quantity[i] * margin[i]

# Print Results to terminal        
        
lines = ['Item-Profit','-----------']
        
for key, value in report.items():
    line = f"{key} {value}"
    lines.append(line)

for line in lines:
    print(line)
    
# Write to PyRamen text file 
with open('PyRamen.txt', 'w') as f:
    for line in lines:
        f.write(line+'\n')
```
