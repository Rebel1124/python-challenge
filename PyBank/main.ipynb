```python
# Import the os and csv library
import os
import csv


# Get csv file path
csvpath = os.getcwd()+'\\Resources\\budget_data.csv'


# Initialize lists to hold data from budget_data.csv
Dates = []
PnL = []


# Import data from budget_data.csv into Dates and PnL lists

with open(csvpath, 'r') as csvfile:
    
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    header = next(csvreader)
    for row in csvreader:
        date = row[0]
        pnl = int(row[1])
        
        Dates.append(date)
        PnL.append(pnl)

## Perform calculations

# Total Months
total_months = len(Dates)

# Total PnL
total = 0
for pnl in PnL:
    total += pnl
    
# Initialize list to hold the change in pnl between dates
pnl_change = []
date_change = Dates[1:]

# Append the pnl changes to the pnl_change list
for i in range(1,len(PnL)):
    change = PnL[i] - PnL[i-1]
    pnl_change.append(change)

    
# Calculate the total pnl change
total_change = 0
for change in pnl_change:
    total_change += change
    
# Using this, we calculate the average change
average_change = round(total_change/len(date_change),2)


# Initialize variables to hold the max_increase (profit) and max_profit_date
max_profit_date = ""
max_increase = 0

# Calulation to workout the max_increase (profit) and max_profit_date
for j in range(0, len(pnl_change)):
    if (pnl_change[j] > max_increase):
        max_increase = pnl_change[j]
        max_profit_date = date_change[j]
        

# Initialize variables to hold the max_decrease (loss) and max_loss_date
max_loss_date = ""
max_decrease = 0

# Calulation to workout the max_decrease (loss) and max_loss_date
for k in range(0, len(pnl_change)):
    if (pnl_change[k] < max_decrease):
        max_decrease = pnl_change[k]
        max_loss_date = date_change[k]
        
# Print Results to terminal
line1 = 'Financial Analysis \n'
line2 = '--------------------- \n'
line3 = 'Total Months: '+str(total_months)+'\n'
line4 = 'Total: ${:0,.0f} \n'.format(total)
line5 = 'Average Change: ${:0,.2f} \n'.format(average_change)
line6 = 'Greatest Increase in Profits: {} ${:0,.0f} \n'.format(max_profit_date, max_increase)
line7 = 'Greatest Decrease in Profits: {} ${:0,.0f} \n'.format(max_loss_date, max_decrease)

lines = [line1, line2, line3, line4, line5, line6, line7]

for line in lines:
    print(line)
    
# Write to PyBank text file   
with open('PyBank.txt', 'w') as f:
    for line in lines:
        f.write(line)
    
```
